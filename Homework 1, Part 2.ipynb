### 2. Monthly closing price plot

```python
import numpy as np
```

```python
sbux_df['Simple return'] = sbux_df['Adj Close'] / sbux_df['Adj Close'].shift(1) - 1
sbux_df['CC return'] = np.log(sbux_df['Adj Close'] / sbux_df['Adj Close'].shift(1))
sbux_df[['Simple return', 'CC return']]
```

```python
plt.plot('Date', 'Simple return', data = sbux_df)
plt.title('Monthly Simple Returns on SBUX')
plt.show()
```

```python
plt.plot('Date', 'CC return', data = sbux_df, color = 'orange')
plt.title('Monthly CC Returns on SBUX')
plt.show()
```

```python
plt.plot('Date', 'Simple return', data = sbux_df, label = 'Simple return')
plt.plot('Date', 'CC return', data = sbux_df, color = 'orange', label = 'CC return')
plt.title('Monthly Returns on SBUX')
plt.legend()
plt.savefig('Return.png')
plt.show()
```
